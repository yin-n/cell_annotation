```python
import json
import pickle
import numpy as np
json_file_path = './test_files.json'

# read test json file
with open(json_file_path) as f:
    test_data = json.load(f)

# # filter norm test data
test_files = [f for f in test_data if '*img.tif' in f]
```

```python
# Define a function to recursively search for matching strings
import json
import fnmatch
def find_matching_strings(obj, pattern):
    matches = []
    if isinstance(obj, dict):
        for key, value in obj.items():
            matches.extend(find_matching_strings(value, pattern))
    elif isinstance(obj, list):
        for item in obj:
            matches.extend(find_matching_strings(item, pattern))
    elif isinstance(obj, str):
        if fnmatch.fnmatch(obj, pattern):
            matches.append(obj)
    return matches

# Define the pattern to match
pattern = '*img.tif'

# Find all matching strings in the JSON data
matching_strings = find_matching_strings(test_data, pattern)
```

```python
test_id = [item.split('_')[0] for item in matching_strings]
test_id = sorted(test_id, key=int)
```

```python
with open('./annotation_position_info.pkl', 'rb') as f:
    ori_data = pickle.load(f)
```

```python
test_df = ori_data[ori_data['ID'].isin(test_id)]
```

```python
random_sample = test_df.sample(n=11, random_state=42)
selected_ids = random_sample.index.tolist()
selected_points = test_df.loc[selected_ids, 'corner'].tolist()
selected_points = np.array(selected_points)
```

```python
remaining_df = test_df.drop(index=selected_ids)
```

```python
remaining_df
```

```python
# Check for overlap using isin()
overlap = remaining_df['ID'].isin(random_sample['ID'])

# Get the IDs that overlap
overlapping_ids = remaining_df[overlap]['ID']

# Check if there is any overlap
if not overlapping_ids.empty:
    print("There is an overlap in the 'ID' column between remaining_df and random_sample.")
    print("Overlapping IDs:")
    print(overlapping_ids)
else:
    print("There is no overlap in the 'ID' column between remaining_df and random_sample.")
```
